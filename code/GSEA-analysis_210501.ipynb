```python
# dataset='norm_GeneMat.txt' #aml
# dataset='norm_lthsc_aavs_mecom_GeneMat.txt' # lthsc
dataset='norm_lthsc_res_mecom_GeneMat.txt'
# dataset='norm_lthsc_aavs_res_GeneMat.txt'

# method="log2_ratio_of_classes"
method="t_test"
# method='signal_to_noise'
# method='ratio_of_classes'

output=exp+'subset/'+dataset[:-4]+"_"+method
aml_norm=pd.read_csv(exp+dataset, sep='\t')
x=[g.split('_')[1] for g in aml_norm.index.to_list()]
aml_norm['Name']=x
# aml_norm['DESCRIPTION']='aml_aavs_vs_mecom'
aml_norm['DESCRIPTION']='aml_mecom_vs_aavs'

gsea_exp=aml_norm[['Name','DESCRIPTION']+aml_norm.columns[:-2].to_list()].reset_index(drop=True)
# gsea_exp=aml_norm[['Name','DESCRIPTION']+aml_norm.columns[::-1][2:].to_list()].reset_index(drop=True)

# enrichr libraries are supported by gsea module. Just provide the name

gs_res = gp.gsea(data=gsea_exp, # or data='./P53_resampling_data.txt'
                 gene_sets='/Users/ltao/Projects/mecom/exp/210210/mecom.gmt', # enrichr library names
                 cls= exp+"lthsc_res_mecom.cls", # cls=class_vector
#                  cls= exp+"lthsc.cls", # cls=class_vector

#                  cls= exp+"aml.cls", # cls=class_vector

                 # set permutation_type to phenotype if samples >=15
                 permutation_type='gene_set',
                 permutation_num=1000, # reduce number to speed up test
                 outdir=output,  # do not write output to disk
                 no_plot=False, # Skip plotting
                 method=method,
                 max_size=1000,
                 processes=4,
                 seed= 7,
                 format='pdf')
```
